## Core Concepts

### 1. Pointers and Dynamic Memory Allocation

*   **Pointers:** Variables that store memory addresses. Essential for direct memory manipulation.
    *   **Declaration:** `int *ptr;`
    *   **Dereferencing:** `*ptr` (access the value at the address `ptr` holds).
    *   **Address-of operator:** `&variable` (get the memory address of `variable`).
*   **Dynamic Memory Allocation:** Allocating memory at runtime, rather than compile time.
    *   **`malloc`/`calloc` (C):** Allocate a block of memory on the heap.
    *   **`free` (C):** Deallocate dynamically allocated memory to prevent memory leaks.
    *   **`new`/`delete` (C++):** Operators for dynamic memory management, often type-safe and integrated with constructors/destructors.
*   **Pointer Arithmetic:** Performing arithmetic operations on pointers (e.g., `ptr++` moves to the next element of the type pointed to).
*   **Common Issues:** Dangling pointers, null pointers, memory leaks, buffer overflows.

### 2. Arrays and Strings (Advanced)

*   **Multi-dimensional Arrays:** Arrays of arrays, useful for matrices or tables.
    *   **Declaration:** `int matrix[ROWS][COLS];`
*   **Arrays of Pointers:** An array where each element is a pointer.
*   **Pointers to Arrays:** A single pointer that points to an entire array.
*   **Strings as Character Arrays:** Understanding how strings are null-terminated character arrays (`char name[] = "Alice";`).
*   **String Manipulation Functions:** Libraries like `string.h` (C) or `<string>` (C++) for operations like copying, concatenation, comparison, and searching.

### 3. Structures and Unions (C) / Structs (C++)

*   **Structures (`struct`):** User-defined data types that group related variables of different types under a single name.
    *   **Declaration:** `struct Person { char name[50]; int age; };`
    *   **Access Members:** `myPerson.age` or `ptrToPerson->age`.
*   **Unions (`union`):** User-defined data types that can hold different data types in the same memory location, but only one member can hold a value at any given time (memory efficient for mutually exclusive data).
*   **Typedef:** Creating aliases for existing data types, often used with structures (`typedef struct Person Person_t;`).

### 4. File Input/Output (I/O)

*   **File Streams:** Representing files as sequences of bytes.
*   **Opening/Closing Files:** Functions like `fopen`/`fclose` (C) or `ifstream`/`ofstream` (C++) to manage file access.
*   **Reading/Writing Data:**
    *   **Formatted I/O:** `fprintf`, `fscanf` (C) for reading/writing formatted text.
    *   **Character I/O:** `fgetc`, `fputc` (C) for single characters.
    *   **Block I/O:** `fread`, `fwrite` (C) for reading/writing raw blocks of data (binary files).
    *   **Stream Operators (C++):** `>>`, `<<` used with `ifstream`/`ofstream`.
*   **Error Handling in File I/O:** Checking for `NULL` file pointers or stream errors.

### 5. Recursion

*   **Definition:** A function that calls itself, either directly or indirectly.
*   **Base Case:** A condition that stops the recursion, preventing infinite loops.
*   **Recursive Step:** The part of the function that makes the recursive call, usually moving closer to the base case.
*   **Applications:** Factorial, Fibonacci sequence, tree traversals, quicksort/mergesort.
*   **Stack Usage:** Understanding how function calls are managed on the call stack during recursion.

### 6. Advanced Control Flow

*   **`switch` statements:** Multi-way branching for specific values.
*   **`break` and `continue`:** Modifying loop behavior.
*   **`goto` statement:** Unconditional jump (generally discouraged due to leading to spaghetti code, but its mechanism is important to understand).

### 7. Preprocessor Directives (C/C++)

*   **`#include`:** Including header files.
*   **`#define`:** Defining macros and symbolic constants.
*   **Conditional Compilation (`#ifdef`, `#ifndef`, `#endif`):** Including or excluding parts of code based on conditions, useful for platform-specific code or debugging.

### 8. Introduction to Data Structures (Specifics vary by course)

*   **Linked Lists:** Dynamic data structures where elements (nodes) are linked together using pointers.
    *   **Types:** Singly, Doubly, Circular.
    *   **Operations:** Insertion, deletion, traversal, searching.

## Functions

Functions are the cornerstone of procedural programming, enabling modularity and reusability.

### 1. Function Prototypes/Declarations

*   Inform the compiler about a function's return type, name, and parameters before its definition.
*   Syntax: `return_type function_name(parameter_type param1, parameter_type param2);`

### 2. Function Definitions

*   The actual implementation of the function's logic.
*   Syntax:
    ```c
    return_type function_name(parameter_type param1, parameter_type param2)
    {
        // Function body
        return value;
    }
    ```

### 3. Parameters and Arguments

*   **Parameters:** Variables declared in the function header that receive values.
*   **Arguments:** The actual values passed to the function when it is called.

### 4. Pass by Value vs. Pass by Reference

*   **Pass by Value:** A copy of the argument's value is passed to the function. Changes to the parameter inside the function do not affect the original argument.
    *   Example: `void func(int x) { x = 10; }`
*   **Pass by Reference (using Pointers):** The memory address of the argument is passed. The function can modify the original argument indirectly using the pointer.
    *   Example: `void func(int *x) { *x = 10; }`
*   **Pass by Reference (using C++ References `&`):** A C++-specific feature where an alias to the original variable is created. Syntactically cleaner than pointers for pass by reference.
    *   Example: `void func(int &x) { x = 10; }`

### 5. Return Values

*   Functions can return a single value of a specified type.
*   `void` return type indicates no value is returned.

### 6. Function Pointers

*   Pointers that store the memory address of a function.
*   Allow functions to be passed as arguments to other functions or stored in data structures.
*   Syntax: `return_type (*function_pointer_name)(parameter_types);`
*   **Applications:** Callbacks, implementing generic algorithms.

### 7. Scope and Storage Classes

*   **Local Scope:** Variables declared inside a function; only accessible within that function.
*   **Global Scope:** Variables declared outside any function; accessible throughout the program.
*   **Static Variables:**
    *   **Local Static:** Retain their value between function calls.
    *   **Global Static:** Limited scope to the file they are declared in (internal linkage).
*   **`extern` Keyword:** Declares that a variable or function is defined elsewhere (external linkage).

### 8. Modular Programming and Header Files

*   **Header Files (`.h` or `.hpp`):** Contain function prototypes, structure definitions, and macro definitions.
*   **Implementation Files (`.c` or `.cpp`):** Contain the actual function definitions.
*   **Purpose:** To separate interfaces from implementations, promote code organization, and allow for easier code reuse and maintenance.
*   **Include Guards:** (`#ifndef`, `#define`, `#endif`) to prevent multiple inclusions of the same header file.

### 1. Pointers and Dynamic Memory Allocation

**Example (C): Dynamic array allocation and deallocation**
```

````python
%%writefile pointer_example.c
#include <stdio.h>
#include <stdlib.h> // For malloc, free

int main() {
    int *arr;
    int n = 5;

    // Allocate memory for 5 integers using malloc
    arr = (int *) malloc(n * sizeof(int));

    // Check if malloc was successful
    if (arr == NULL) {
        printf("Memory allocation failed!\n");
        return 1;
    }

    // Initialize and print the array elements
    printf("Dynamically allocated array:\n");
    for (int i = 0; i < n; i++) {
        arr[i] = i * 10;
        printf("arr[%d] = %d (Address: %p)\n", i, arr[i], (void*)(arr + i));
    }

    // Free the allocated memory
    free(arr);
    printf("Memory freed.\n");

    // Attempting to access after free can lead to undefined behavior
    // printf("After free: arr[0] = %d\n", arr[0]); // DO NOT DO THIS IN REAL CODE

    return 0;
}
````

### 5. Recursion

**Example (C): Factorial calculation using recursion**
```

````python
%%writefile recursion_example.c
#include <stdio.h>

// Recursive function to calculate factorial
long long factorial(int n) {
    // Base case: if n is 0 or 1, factorial is 1
    if (n == 0 || n == 1) {
        return 1;
    }
    // Recursive step: n * factorial(n-1)
    else {
        return n * factorial(n - 1);
    }
}

int main() {
    int num = 5;
    printf("Factorial of %d is %lld\n", num, factorial(num)); // Expected: 120

    num = 0;
    printf("Factorial of %d is %lld\n", num, factorial(num)); // Expected: 1

    return 0;
}
````

### 6. Advanced Control Flow

**Example (C): Using `switch` and `break`/`continue`**
```

````python
%%writefile control_flow_example.c
#include <stdio.h>

int main() {
    int choice = 2;

    // --- switch statement ---
    switch (choice) {
        case 1:
            printf("Choice is 1\n");
            break; // Exits the switch
        case 2:
            printf("Choice is 2\n");
            // Fallthrough if no break here
        case 3:
            printf("Choice is 3 (or fell through from 2)\n");
            break;
        default:
            printf("Invalid choice\n");
            break;
    }

    printf("\nDemonstrating continue and break in a loop:\n");
    for (int i = 0; i < 5; i++) {
        if (i == 2) {
            printf("Skipping iteration %d with continue\n", i);
            continue; // Skips the rest of the current iteration
        }
        if (i == 4) {
            printf("Breaking loop at iteration %d\n", i);
            break; // Exits the loop entirely
        }
        printf("Current i: %d\n", i);
    }
    printf("Loop finished.\n");

    return 0;
}
````

### 8. Introduction to Data Structures: Linked Lists

**Example (C): Simple singly linked list node and insertion**
```

````python
%%writefile linked_list_example.c
#include <stdio.h>
#include <stdlib.h> // For malloc, free

// Define a node structure for a singly linked list
struct Node {
    int data;
    struct Node *next; // Pointer to the next node
};

// Function to insert a new node at the beginning of the list
struct Node* insertAtBeginning(struct Node *head, int value) {
    struct Node *newNode = (struct Node*) malloc(sizeof(struct Node));
    if (newNode == NULL) {
        printf("Memory allocation failed!\n");
        exit(1);
    }
    newNode->data = value;
    newNode->next = head; // New node points to the current head
    return newNode;        // New node becomes the head
}

// Function to print the linked list
void printList(struct Node *head) {
    struct Node *current = head;
    printf("Linked List: ");
    while (current != NULL) {
        printf("%d -> ", current->data);
        current = current->next;
    }
    printf("NULL\n");
}

// Function to free the memory of the linked list
void freeList(struct Node *head) {
    struct Node *current = head;
    struct Node *next_node;
    while (current != NULL) {
        next_node = current->next;
        free(current);
        current = next_node;
    }
    printf("Linked list memory freed.\n");
}

int main() {
    struct Node *head = NULL; // Initialize an empty list

    head = insertAtBeginning(head, 30);
    head = insertAtBeginning(head, 20);
    head = insertAtBeginning(head, 10);

    printList(head);

    freeList(head);

    return 0;
}
````

### 9. Introduction to Algorithms: Linear Search

**Example (C): Implementing a linear search algorithm**
```

````python
%%writefile linear_search_example.c
#include <stdio.h>

// Function to perform linear search
int linearSearch(int arr[], int size, int target) {
    for (int i = 0; i < size; i++) {
        if (arr[i] == target) {
            return i; // Return the index if target is found
        }
    }
    return -1; // Return -1 if target is not found
}

int main() {
    int numbers[] = {10, 20, 30, 40, 50};
    int size = sizeof(numbers) / sizeof(numbers[0]);
    int target1 = 30;
    int target2 = 99;

    int result1 = linearSearch(numbers, size, target1);
    if (result1 != -1) {
        printf("Target %d found at index %d.\n", target1, result1);
    } else {
        printf("Target %d not found.\n", target1);
    }

    int result2 = linearSearch(numbers, size, target2);
    if (result2 != -1) {
        printf("Target %d found at index %d.\n", target2, result2);
    } else {
        printf("Target %d not found.\n", target2);
    }

    return 0;
}
````

### 2. Pass by Value vs. Pass by Reference

**Example (C): Demonstrating pass by value and pass by reference (using pointers)**
```

````python
%%writefile pass_by_example.c
#include <stdio.h>

// Function demonstrating pass by value
void incrementByValue(int num) {
    printf("Inside incrementByValue (before): num = %d\n", num);
    num++; // Increments a copy of the argument
    printf("Inside incrementByValue (after): num = %d\n", num);
}

// Function demonstrating pass by reference using a pointer
void incrementByReference(int *numPtr) {
    printf("Inside incrementByReference (before): *numPtr = %d\n", *numPtr);
    (*numPtr)++; // Increments the value at the memory address pointed to
    printf("Inside incrementByReference (after): *numPtr = %d\n", *numPtr);
}

int main() {
    int value = 10;
    printf("Initial value: %d\n", value);

    printf("\nCalling incrementByValue:\n");
    incrementByValue(value);
    printf("After incrementByValue: value = %d\n", value); // Value remains unchanged

    printf("\nCalling incrementByReference:\n");
    incrementByReference(&value); // Pass the address of 'value'
    printf("After incrementByReference: value = %d\n", value); // Value is changed

    return 0;
}
````

### 3. Function Pointers

**Example (C): Using function pointers**
```

````python
%%writefile func_pointer_example.c
#include <stdio.h>

// Simple functions
int add(int a, int b) {
    return a + b;
}

int subtract(int a, int b) {
    return a - b;
}

// Function that takes a function pointer as an argument
int operate(int x, int y, int (*operation)(int, int)) {
    return operation(x, y);
}

int main() {
    int result;

    // Declare a function pointer
    int (*func_ptr)(int, int);

    // Point to the add function
    func_ptr = add;
    result = func_ptr(10, 5);
    printf("Using func_ptr to add: %d\n", result); // Expected: 15

    // Point to the subtract function
    func_ptr = subtract;
    result = func_ptr(10, 5);
    printf("Using func_ptr to subtract: %d\n", result); // Expected: 5

    // Using operate function with different operations
    printf("Using operate with add: %d\n", operate(20, 8, add)); // Expected: 28
    printf("Using operate with subtract: %d\n", operate(20, 8, subtract)); // Expected: 12

    return 0;
}
````

````python
%%writefile main.c
#include <stdio.h>
#include "my_math.h" // Include our custom header

int main() {
    int x = 10, y = 5;

    printf("Addition: %d + %d = %d\n", x, y, add(x, y));
    printf("Subtraction: %d - %d = %d\n", x, y, subtract(x, y));

    return 0;
}
````
