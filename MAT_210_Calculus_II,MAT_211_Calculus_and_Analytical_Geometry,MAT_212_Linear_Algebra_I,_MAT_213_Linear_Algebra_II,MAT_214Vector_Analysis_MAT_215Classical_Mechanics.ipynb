# Worked Examples in Python for Math and Physics

This notebook provides worked examples in Python for concepts related to Vector Analysis and Classical Mechanics, which are typically covered in courses like MAT 214 and MAT 215.

## Vector Analysis (MAT 214)

```python
import numpy as np

# Example 1: Dot Product of two vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

dot_product = np.dot(vector_a, vector_b)

print(f"Vector A: {vector_a}")
print(f"Vector B: {vector_b}")
print(f"Dot product of A and B: {dot_product}")
```

### Explanation: Dot Product
The dot product (or scalar product) of two vectors results in a single scalar value. It's often used to find the angle between two vectors or to project one vector onto another. In `numpy`, the `np.dot()` function efficiently calculates the dot product.

```python
# Example 2: Cross Product of two 3D vectors
vector_c = np.array([1, 0, 0])  # Unit vector along x-axis
vector_d = np.array([0, 1, 0])  # Unit vector along y-axis

cross_product = np.cross(vector_c, vector_d)

print(f"Vector C: {vector_c}")
print(f"Vector D: {vector_d}")
print(f"Cross product of C and D: {cross_product}")
```

### Explanation: Cross Product
The cross product (or vector product) of two vectors in 3D space results in a new vector that is perpendicular to both original vectors. Its magnitude is related to the area of the parallelogram formed by the two vectors. `np.cross()` is used for this calculation in `numpy`.

## Classical Mechanics (MAT 215)

```python
# Example: Calculating Displacement and Final Velocity for Uniform Acceleration

# Given values
initial_velocity = 10  # m/s
acceleration = 2     # m/s^2
time = 5             # seconds

# Calculate final velocity (v = u + at)
final_velocity = initial_velocity + (acceleration * time)

# Calculate displacement (s = ut + 0.5at^2)
displacement = (initial_velocity * time) + (0.5 * acceleration * (time**2))

print(f"Initial Velocity (u): {initial_velocity} m/s")
print(f"Acceleration (a): {acceleration} m/s^2")
print(f"Time (t): {time} s")
print(f"\nCalculated Final Velocity (v): {final_velocity} m/s")
print(f"Calculated Displacement (s): {displacement} m")
```

### Explanation: Uniform Acceleration
This example demonstrates how to use fundamental kinematic equations to calculate final velocity and displacement for an object moving with constant (uniform) acceleration. The formulas used are:
*   **Final Velocity**: `v = u + at`
*   **Displacement**: `s = ut + 0.5at^2`

Where:
*   `u` is initial velocity
*   `v` is final velocity
*   `a` is acceleration
*   `t` is time
*   `s` is displacement
