```python
# STEP 1: Install dependencies (Colab)
!pip install -q langchain langchain-community faiss-cpu sentence-transformers transformers accelerate umap-learn pypdf matplotlib
!pip install requests==2.32.4 --force-reinstall

# STEP 2: Imports
import os
import pickle
import glob
import matplotlib.pyplot as plt
import umap
from google.colab import files  # remove if running locally

from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

# --------- CONFIG ----------
PDF_FOLDER = "uploaded_pdfs"   # folder to place PDFs
EMBED_MODEL = "sentence-transformers/all-MiniLM-L6-v2"
LLM_MODEL = "google/flan-t5-base"
CHUNK_SIZE = 1000
CHUNK_OVERLAP = 200
K_RETRIEVE = 3
# ---------------------------

# STEP 3: Upload multiple PDFs (Colab)
print("📂 If using Colab: upload one or more PDFs in the chooser that appears.")
uploaded = files.upload()  # comment out if running locally
os.makedirs(PDF_FOLDER, exist_ok=True)
for fname in uploaded.keys():
    dest = os.path.join(PDF_FOLDER, fname)
    with open(dest, "wb") as f:
        f.write(uploaded[fname])
print("✅ Uploaded files:", list(uploaded.keys()))

# STEP 4: Load all PDFs
all_documents = []
for pdf_file in glob.glob(os.path.join(PDF_FOLDER, "*.pdf")):
    loader = PyPDFLoader(pdf_file)
    all_documents.extend(loader.load())

print(f"📂 Loaded {len(all_documents)} total pages from {len(uploaded)} PDFs")

# STEP 5: Split into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)
docs = text_splitter.split_documents(all_documents)

# Attach fallback metadata if missing
for d in docs:
    d.metadata.setdefault("source", "unknown")
    d.metadata.setdefault("page", -1)

print(f"📑 Created {len(docs)} chunks from all PDFs.")
if docs:
    print("\n--- Sample chunk (with metadata) ---")
    print(docs[0].page_content[:300])
    print("metadata:", docs[0].metadata)
else:
    print("⚠️ No chunks created. Check if PDFs are empty or scanned images.")

# STEP 6: Embeddings + FAISS (HuggingFace local)
if docs:
    embeddings = HuggingFaceEmbeddings(model_name=EMBED_MODEL)
    vectorstore = FAISS.from_documents(docs, embeddings)
    vectorstore.save_local("faiss_store")

    # Also prepare raw vectors for visualization
    doc_texts = [d.page_content[:200] for d in docs]
    doc_vectors = embeddings.embed_documents([d.page_content for d in docs])
    print("✅ FAISS vector store built and saved locally.")

    # STEP 7: UMAP visualization of embeddings
    reducer = umap.UMAP(n_neighbors=10, min_dist=0.1, random_state=42)
    embedding_2d = reducer.fit_transform(doc_vectors)

    plt.figure(figsize=(10, 7))
    plt.scatter(embedding_2d[:, 0], embedding_2d[:, 1], c="blue", alpha=0.6)
    for i in range(0, len(doc_texts), max(1, len(doc_texts)//15)):
        plt.text(embedding_2d[i, 0], embedding_2d[i, 1], doc_texts[i][:30] + "...", fontsize=8)
    plt.title("📊 PDF Chunks in Embedding Space")
    plt.show()

    # STEP 8: Load local LLM (HuggingFace) + RetrievalQA
    generator = pipeline(
        "text2text-generation",
        model=LLM_MODEL,
        device_map="auto",
        max_new_tokens=256
    )
    llm = HuggingFacePipeline(pipeline=generator)

    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        retriever=vectorstore.as_retriever(search_kwargs={"k": K_RETRIEVE}),
        chain_type="stuff"
    )
    print("✅ RetrievalQA ready (HuggingFace LLM).")

    # STEP 9: Retrieval visualization function
    def visualize_retrieval(query, retriever, all_docs, all_vectors, reducer):
        retrieved = retriever.get_relevant_documents(query)
        retrieved_texts = [doc.page_content for doc in retrieved]
        retrieved_meta = [doc.metadata for doc in retrieved]

        retrieved_indices = []
        for r in retrieved:
            for i, d in enumerate(all_docs):
                if d.page_content == r.page_content:
                    retrieved_indices.append(i)
                    break

        embedding_2d = reducer.transform(all_vectors)

        plt.figure(figsize=(10, 7))
        plt.scatter(embedding_2d[:, 0], embedding_2d[:, 1], c="lightgray", alpha=0.5, label="All Chunks")
        if retrieved_indices:
            plt.scatter(embedding_2d[retrieved_indices, 0], embedding_2d[retrieved_indices, 1],
                        c="red", alpha=0.9, label="Retrieved Chunks")
        plt.title(f"🔎 Retrieved Chunks for Query:\n'{query}'")
        plt.legend()
        plt.show()

        for i, (text, meta) in enumerate(zip(retrieved_texts, retrieved_meta), 1):
            print(f"\n--- Retrieved Chunk {i} ---\nSource: {meta.get('source')}  Page: {meta.get('page')}\n{text[:400]}...")

    # Example test
    query = "What is the main topic of these documents?"
    answer = qa_chain.run(query)
    print("Q:", query)
    print("A:", answer)
    visualize_retrieval(query, qa_chain.retriever, docs, doc_vectors, reducer)

    # STEP 10: Interactive loop
    print("\n🤖 Interactive multi-PDF Q&A (type 'exit' to quit)")
    while True:
        q = input("\nAsk a question: ")
        if q.strip().lower() in ["exit", "quit"]:
            print("👋 Goodbye.")
            break
        try:
            answer = qa_chain.run(q)
            print("\n💡 A:", answer)
            visualize_retrieval(q, qa_chain.retriever, docs, doc_vectors, reducer)
        except Exception as e:
            print("⚠️ Error:", e)
else:
    print("Skipping embedding and QA steps because no document chunks were created.")
```
