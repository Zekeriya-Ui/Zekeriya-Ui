### 1. Area of a Rectangle

The area of a rectangle is calculated by multiplying its length by its width.

Formula: `Area = length * width`

````python
# Function to calculate the area of a rectangle
def calculate_rectangle_area(length, width):
    return length * width

# Example usage
length = 10
width = 5
area = calculate_rectangle_area(length, width)
print(f"The area of a rectangle with length {length} and width {width} is: {area}")
````

### 2. Area of a Circle

The area of a circle is calculated using the formula `πr²`, where `r` is the radius and `π` (pi) is a mathematical constant approximately equal to 3.14159.

Formula: `Area = π * radius²`

````python
import math

# Function to calculate the area of a circle
def calculate_circle_area(radius):
    return math.pi * (radius ** 2)

# Example usage
radius = 7
area = calculate_circle_area(radius)
print(f"The area of a circle with radius {radius} is: {area:.2f}")
````

### 3. Distance Between Two Points

The distance between two points (x1, y1) and (x2, y2) in a 2D plane can be found using the distance formula, derived from the Pythagorean theorem.

Formula: `Distance = √((x2 - x1)² + (y2 - y1)²) `

````python
import math

# Function to calculate the distance between two points
def calculate_distance(x1, y1, x2, y2):
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

# Example usage
x1, y1 = 1, 2
x2, y2 = 4, 6
distance = calculate_distance(x1, y1, x2, y2)
print(f"The distance between ({x1}, {y1}) and ({x2}, {y2}) is: {distance:.2f}")
````

### 1. Introduction to Derivatives

A derivative measures the sensitivity of change of the function value (output value) with respect to a change in its argument (input value). It represents the instantaneous rate of change or the slope of the tangent line to the curve at a given point.

We'll use the `sympy` library to symbolically calculate derivatives.

````python
from sympy import symbols, diff, sin

# Define a symbolic variable
x = symbols('x')

# Define a function
f_x = x**2 + 3*x + 2

# Calculate the derivative of the function with respect to x
derivative_f_x = diff(f_x, x)

print(f"Original function: f(x) = {f_x}")
print(f"Derivative: f'(x) = {derivative_f_x}")

# Another example: derivative of sin(x)
f_sin_x = symbols('x')
derivative_sin_x = diff(sin(f_sin_x), f_sin_x)
print(f"Original function: f(x) = sin(x)")
print(f"Derivative: f'(x) = {derivative_sin_x}")
````

### 2. Introduction to Integration

Integration is the reverse process of differentiation. It is used to find the total sum or accumulation of a quantity, often representing the area under a curve. An indefinite integral (antiderivative) results in a family of functions, while a definite integral results in a specific numerical value.

We'll use the `sympy` library to symbolically calculate integrals.

````python
from sympy import symbols, integrate

# Define a symbolic variable
x = symbols('x')

# Define a function
g_x = 2*x + 3

# Calculate the indefinite integral of the function with respect to x
indefinite_integral_g_x = integrate(g_x, x)

print(f"Original function: g(x) = {g_x}")
print(f"Indefinite Integral: ∫g(x)dx = {indefinite_integral_g_x} + C")

# Calculate the definite integral from 0 to 2
definite_integral_g_x = integrate(g_x, (x, 0, 2))
print(f"Definite Integral from 0 to 2: ∫[0,2]g(x)dx = {definite_integral_g_x}")
````

### 3. Derivative of a Complex Function with SymPy

SymPy can handle more complex functions involving products, quotients, and chain rule. You simply define your function, and `diff()` will apply the necessary rules automatically.

Let's consider an example function: `f(x) = (x^2 + 1) * sin(3x)`.
This function requires both the product rule and the chain rule for differentiation.

````python
from sympy import symbols, diff, sin, cos

# Define the symbolic variable
x = symbols('x')

# Define the complex function
f_complex = (x**2 + 1) * sin(3*x)

# Calculate the derivative
derivative_f_complex = diff(f_complex, x)

print(f"Original complex function: f(x) = {f_complex}")
print(f"Derivative: f'(x) = {derivative_f_complex}")

# To verify, let's manually apply the rules:
# Product Rule: (uv)' = u'v + uv'
# u = x^2 + 1 => u' = 2x
# v = sin(3x) => v' = cos(3x) * 3 (Chain Rule)
# So, f'(x) = (2x) * sin(3x) + (x^2 + 1) * (3 * cos(3x))
# f'(x) = 2x * sin(3x) + 3 * (x^2 + 1) * cos(3x)

# SymPy simplifies this automatically.
````

### 4. Explanation of the Product Rule

The **Product Rule** is a formula used to find the derivative of a product of two or more functions. If you have two differentiable functions, say `u(x)` and `v(x)`, then the derivative of their product `P(x) = u(x) * v(x)` is given by:

`P'(x) = u'(x) * v(x) + u(x) * v'(x)`

In words, it's "the derivative of the first function times the second function, plus the first function times the derivative of the second function."

#### Application in the previous example:

For the function `f(x) = (x^2 + 1) * sin(3x)`:

*   Let `u(x) = x^2 + 1`
*   Let `v(x) = sin(3x)`

First, find the derivatives of `u(x)` and `v(x)`:

*   `u'(x) = d/dx (x^2 + 1) = 2x`
*   `v'(x) = d/dx (sin(3x))`
    *   Here, we need the **Chain Rule** because `sin(3x)` is a composite function. The derivative of `sin(g(x))` is `cos(g(x)) * g'(x)`. So, `d/dx (sin(3x)) = cos(3x) * d/dx (3x) = cos(3x) * 3 = 3cos(3x)`.

Now, apply the product rule formula:

`f'(x) = u'(x) * v(x) + u(x) * v'(x)`
`f'(x) = (2x) * sin(3x) + (x^2 + 1) * (3cos(3x))`
`f'(x) = 2x sin(3x) + 3(x^2 + 1) cos(3x)`

This matches the result SymPy provided, demonstrating the automatic application of these rules.

### 5. Explanation of the Quotient Rule

The **Quotient Rule** is used to find the derivative of a function that is the ratio of two other differentiable functions. If you have a function `Q(x) = u(x) / v(x)`, where `v(x) ≠ 0`, then the derivative of `Q(x)` is given by:

`Q'(x) = (u'(x) * v(x) - u(x) * v'(x)) / (v(x))²`

In words, it's "the derivative of the top function times the bottom function, minus the top function times the derivative of the bottom function, all divided by the bottom function squared."

Let's calculate the derivative of a function using the quotient rule with SymPy.

````python
from sympy import symbols, diff

# Define the symbolic variable
x = symbols('x')

# Define a function that requires the quotient rule: f(x) = (x^2) / (x + 1)
u_x = x**2
v_x = x + 1
f_quotient = u_x / v_x

# Calculate the derivative using SymPy
derivative_f_quotient = diff(f_quotient, x)

print(f"Original function: f(x) = {f_quotient}")
print(f"Derivative using SymPy: f'(x) = {derivative_f_quotient}")

# To verify, let's manually apply the quotient rule:
# u(x) = x^2  => u'(x) = 2x
# v(x) = x + 1 => v'(x) = 1
# Q'(x) = (u'v - uv') / v^2
# Q'(x) = (2x * (x + 1) - x^2 * 1) / (x + 1)^2
# Q'(x) = (2x^2 + 2x - x^2) / (x + 1)^2
# Q'(x) = (x^2 + 2x) / (x + 1)^2
````

````python
from sympy import simplify, Eq

# The derivative obtained from SymPy (from previous cell)
sympy_derivative = derivative_f_quotient

# The manually simplified derivative
manual_derivative = (x**2 + 2*x) / (x + 1)**2

print(f"SymPy's derivative: {sympy_derivative}")
print(f"Manual simplified derivative: {manual_derivative}")

# Check if they are algebraically equivalent after simplification
# SymPy automatically simplifies expressions when comparing for equality with `Eq` or `simplify`
if simplify(sympy_derivative) == simplify(manual_derivative):
    print("\nBoth derivatives are algebraically equivalent!")
else:
    print("\nThe derivatives are NOT algebraically equivalent.")

# Another way to check for equality is to see if their difference simplifies to zero
if simplify(sympy_derivative - manual_derivative) == 0:
    print("The difference between the derivatives simplifies to zero, confirming equivalence.")
else:
    print("The difference between the derivatives does not simplify to zero.")
````

### 6. Introduction to Limits

A **limit** in calculus describes the behavior of a function as its input (variable) gets arbitrarily close to a certain value. It does not necessarily evaluate the function at that exact point, but rather describes the value that the function \"approaches\" as the input gets closer and closer. Limits are fundamental to calculus and are used to define continuity, derivatives, and integrals.

Mathematically, the expression $\lim_{x \to c} f(x) = L$ means that as $x$ approaches $c$ (from either side), the value of $f(x)$ approaches $L$. This concept is crucial for understanding rates of change and accumulation.

### 7. Introduction to Continuity

A function is **continuous** at a particular point if its graph has no breaks, jumps, or holes at that point. Intuitively, you can draw the graph of a continuous function without lifting your pen from the paper.

More formally, a function $f(x)$ is continuous at a point $c$ if the following three conditions are met:

1.  **The function is defined at c**: $f(c)$ exists.
2.  **The limit of the function exists at c**: $\lim_{x \to c} f(x)$ exists.
3.  **The limit equals the function's value at c**: $\lim_{x \to c} f(x) = f(c)$.

If any of these conditions are not met, the function is discontinuous at $c$. Understanding continuity is crucial for many theorems in calculus, such as the Intermediate Value Theorem and the Extreme Value Theorem.

### 8. Applications of Derivatives: Optimization

Derivatives are powerful tools for solving optimization problems, which involve finding the maximum or minimum value of a function. This is particularly useful in many real-world scenarios, such as maximizing profit, minimizing cost, or finding the most efficient design.

#### Key Concepts:

1.  **Critical Points**: These are the points where the first derivative of a function is either equal to zero (`f'(x) = 0`) or undefined. Critical points are candidates for local maxima, local minima, or saddle points.

2.  **Local Maxima and Minima**: These are points where the function reaches its highest or lowest value within a certain interval. At a local maximum, the function changes from increasing to decreasing, and at a local minimum, it changes from decreasing to increasing.

3.  **Second Derivative Test**: This test helps classify critical points:
    *   If `f''(c) < 0` at a critical point `c`, then `f(c)` is a **local maximum**.
    *   If `f''(c) > 0` at a critical point `c`, then `f(c)` is a **local minimum**.
    *   If `f''(c) = 0`, the test is inconclusive, and one might need to use the First Derivative Test or further analysis.

#### How it Works:

To find the local extrema of a function `f(x)`:
1.  Find the first derivative, `f'(x)`.
2.  Set `f'(x) = 0` and solve for `x` to find the critical points.
3.  Find the second derivative, `f''(x)`.
4.  Substitute each critical point into `f''(x)` to apply the Second Derivative Test.
    *   If `f''(c) < 0`, it's a local maximum.
    *   If `f''(c) > 0`, it's a local minimum.
    *   If `f''(c) = 0`, the test doesn't provide a clear answer.

````python
from sympy import symbols, diff, solve

# Define a symbolic variable
x = symbols('x')

# Define the function for optimization example
f_optimization = x**3 - 6*x**2 + 9*x + 1

print(f"Original function: f(x) = {f_optimization}")

# 1. Calculate the first derivative
first_derivative = diff(f_optimization, x)
print(f"\nFirst Derivative: f'(x) = {first_derivative}")

# 2. Find critical points by setting the first derivative to zero
critical_points = solve(first_derivative, x)
print(f"\nCritical points (where f'(x) = 0): {critical_points}")

# 3. Calculate the second derivative
second_derivative = diff(f_optimization, x, 2)
print(f"\nSecond Derivative: f''(x) = {second_derivative}")

# 4. Classify critical points using the Second Derivative Test
print("\nClassification of Critical Points (using Second Derivative Test):")
for point in critical_points:
    second_derivative_at_point = second_derivative.subs(x, point)
    if second_derivative_at_point < 0:
        print(f"  At x = {point}: f''({point}) = {second_derivative_at_point} < 0 => Local Maximum")
    elif second_derivative_at_point > 0:
        print(f"  At x = {point}: f''({point}) = {second_derivative_at_point} > 0 => Local Minimum")
    else:
        print(f"  At x = {point}: f''({point}) = {second_derivative_at_point} = 0 => Second Derivative Test is inconclusive")
````

### 9. Applications of Integrals: Area Under a Curve

One of the most fundamental applications of definite integrals is to calculate the **area under a curve**. If a function `f(x)` is non-negative over an interval `[a, b]`, the definite integral of `f(x)` from `a` to `b` represents the area of the region bounded by the graph of `f(x)`, the x-axis, and the vertical lines `x = a` and `x = b`.

Formula: `Area = ∫[a,b] f(x) dx`

#### Key Concepts:

1.  **Definite Integral**: Unlike indefinite integrals (antiderivatives), a definite integral evaluates the integral over a specific interval, resulting in a numerical value.
2.  **Limits of Integration**: These are the values `a` (lower limit) and `b` (upper limit) that define the interval over which the area is calculated.
3.  **Geometric Interpretation**: The definite integral can be thought of as summing up infinitely many infinitesimally thin rectangles under the curve.

Let's use SymPy to calculate the area under a curve for a given function and interval, and then visualize it using `matplotlib`.

**Reasoning**:
Now that the concept of applications of integrals for calculating the area under a curve has been explained in a markdown block, I will add a new code cell to demonstrate how to calculate definite integrals using SymPy and visualize the area using Matplotlib, as outlined in the instructions.



````python
from sympy import symbols, integrate, lambdify
import numpy as np
import matplotlib.pyplot as plt

# Define a symbolic variable
x = symbols('x')

# Define a function to find the area under
f_area = x**2 + 1

# Define the limits of integration
a = 0
b = 2

print(f"Original function: f(x) = {f_area}")
print(f"Limits of integration: from x = {a} to x = {b}")

# Calculate the definite integral (Area Under Curve) using SymPy
area_under_curve = integrate(f_area, (x, a, b))
print(f"\nArea under the curve: {area_under_curve}")

# --- Visualization ---

# Generate x values for plotting
x_vals = np.linspace(a - 1, b + 1, 400) # Extend range slightly for better visualization

# Convert sympy function to a numerical function for plotting
f_area_numeric = lambdify(x, f_area, 'numpy')
y_vals_f_area = f_area_numeric(x_vals)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_f_area, label=f'f(x) = {f_area}', color='blue')

# Shade the area under the curve between a and b
x_fill = np.linspace(a, b, 100)
y_fill = f_area_numeric(x_fill)
plt.fill_between(x_fill, y_fill, color='lightgray', alpha=0.7, label=f'Area from {a} to {b}')

# Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title(f'Area Under the Curve for f(x) = {f_area} from {a} to {b}')

# Add a legend
plt.legend()

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
````

````python
from sympy import symbols, integrate, diff

# Define a symbolic variable
x = symbols('x')

# Define a function
f_ftc = 2*x + 3

# Define limits of integration
a = 0
b = 2

print(f"Function: f(x) = {f_ftc}")
print(f"Limits of integration: from x = {a} to x = {b}")

# --- Part 2 of FTC: Evaluate definite integral using antiderivative ---

# 1. Find the indefinite integral (antiderivative)
F_x = integrate(f_ftc, x)
print(f"\nAntiderivative: F(x) = {F_x} + C")

# 2. Evaluate F(b) - F(a)
definite_integral_ftc = F_x.subs(x, b) - F_x.subs(x, a)
print(f"Value of F(b) - F(a) = F({b}) - F({a}) = {definite_integral_ftc}")

# 3. Verify with direct definite integral calculation
direct_definite_integral = integrate(f_ftc, (x, a, b))
print(f"Direct definite integral calculation: {direct_definite_integral}")

if definite_integral_ftc == direct_definite_integral:
    print("\nBoth methods yield the same result, confirming FTC Part 2.")
else:
    print("\nThere is a discrepancy in the calculations.")

# --- Part 1 of FTC (demonstration concept) ---
# If F(x) = integral from 'a' to 'x' of f(t) dt, then F'(x) = f(x)
# Let's consider F(x) = integrate(f_ftc, (x, a, x))
F_of_x = integrate(f_ftc, (x, a, x))
print(f"\nLet G(x) = integral from {a} to x of f(t) dt = {F_of_x}")

derivative_of_F_of_x = diff(F_of_x, x)
print(f"Then G'(x) = {derivative_of_F_of_x}")

if derivative_of_F_of_x == f_ftc:
    print("This confirms FTC Part 1: G'(x) equals f(x).")
else:
    print("Discrepancy in FTC Part 1 demonstration.")
````

### 10. The Fundamental Theorem of Calculus (FTC)

The Fundamental Theorem of Calculus is a cornerstone of calculus, linking the concepts of differentiation and integration. It consists of two parts.

#### Part 1: The First Fundamental Theorem of Calculus
This part states that if $f$ is continuous on an interval $[a, b]$, then the function $F(x)$ defined as the integral of $f$ from $a$ to $x$ is an antiderivative of $f$. In other words, differentiation and integration are inverse operations.

Formula: $F(x) = \int_{a}^{x} f(t) dt \implies F'(x) = f(x)$

#### Part 2: The Second Fundamental Theorem of Calculus
This is the more commonly used part for evaluating definite integrals. It states that if $F$ is any antiderivative of a continuous function $f$ on $[a, b]$, then the definite integral of $f$ from $a$ to $b$ can be evaluated by simply finding an antiderivative of $f$ and evaluating it at the limits of integration.

Formula: $\int_{a}^{b} f(x) dx = F(b) - F(a)$

Where $F'(x) = f(x)$. This theorem simplifies the calculation of definite integrals, avoiding the need to compute limits of Riemann sums.
